#### P3. 1. Creating Gradient Descent Algorithm

```python
def gradient (filtered_m,X):
    maximum_count = 10000
    count = 0
    es = []
    error_less = False
    while not error_less:
        
        filtered_m_Transpose = np.transpose(filtered_m)
        filter_X_multiply = np.matmul(filtered_m_Transpose,X)
        g_x = np.divide(1,1+np.exp(-(filter_X_multiply)))
        
        g_dash_x = np.multiply(g_x,np.subtract(1,g_x))
        
        S_min_logistic_function = np.subtract(sg_matrix,g_x)
        
        
        S_min_logistic_function_Transpose = np.transpose(S_min_logistic_function)
        
        
        error = ((np.matmul(S_min_logistic_function,S_min_logistic_function_Transpose)/sg_matrix.size)[0][0])
        
        
        dE_df = 2*np.matmul(X,np.transpose((np.multiply(S_min_logistic_function,g_dash_x))))/sg_matrix.size
        
        filtered_m = filtered_m + np.multiply(0.05,dE_df)
        
        if np.abs(error) <= 0.01:
            error_less = True
        es.append(error)
        
        if count == maximum_count:
            error_less = True
        count += 1
         

    return filtered_m_Transpose,es
        
```

Creating BFGS algorithm

```python
def BFGS(filtered_m,X):
    maximum_count = 1000
    count = 0
    es = []
    error_less = False
    g = np.eye(9)*0.1
    while not error_less:
        filtered_m_Transpose = np.transpose(filtered_m)
        filter_X_multiply = np.matmul(filtered_m_Transpose,X)
        g_x = np.divide(1,1+np.exp(-(filter_X_multiply)))
        
        g_dash_x = np.multiply(g_x,np.subtract(1,g_x))
        
        S_min_logistic_function = np.subtract(sg_matrix,g_x)
        
        
        S_min_logistic_function_Transpose = np.transpose(S_min_logistic_function)
        
        error = ((np.matmul(S_min_logistic_function,S_min_logistic_function_Transpose)/sg_matrix.size)[0][0])

        
        if count == 0:
            dE_df = 2*np.matmul(X,np.transpose((np.multiply(S_min_logistic_function,g_dash_x))))/sg_matrix.size
            G = g
            dE_df_plus1 =  dE_df
            filtered_m_plus1 = filtered_m
            filtered_m = filtered_m - np.matmul(g,dE_df)        
        else:
            dE_df_plus1 =  dE_df

            dE_df = -2*np.matmul(X,np.transpose((np.multiply(S_min_logistic_function,g_dash_x))))/sg_matrix.size
            
            p = filtered_m_plus1 - filtered_m
            p_transpose = np.transpose(p)
            
            v = dE_df_plus1 - dE_df
            v_transpose = np.transpose(v)
            
            u = np.subtract(np.divide(p,np.matmul(p_transpose,v)),np.divide(np.matmul(g,v),np.matmul(v_transpose,np.matmul(g,v))))
            u_transpose = np.transpose(u)
            
            a = g
            b = np.divide(np.matmul(p,p_transpose),np.matmul(p_transpose,v))
            c = np.divide(np.matmul(np.matmul(np.matmul(g,v),v_transpose),g),np.matmul(np.matmul(v_transpose,g),v))
            d = (np.matmul(v_transpose,np.matmul(g,v))*np.matmul(u,u_transpose))
            G = g
            G = a + b - c + d
            filtered_m_plus1 = filtered_m

            filtered_m = filtered_m - np.matmul(g,dE_df)        



            
        if np.abs(error) <= 0.01:
            error_less = True
        es.append(error)
        
        if count == maximum_count:
            error_less = True
        count += 1
         

    return filtered_m_Transpose,es  
```

#### P5. 5. Kmeans clustering

```python
def Kmeans(matrix):
    random_centres = random.sample(list(set(matrix)),4)
    while True:
        centroids = [[],[],[],[]]
        centres = [[],[],[],[]]
        for i in (matrix):
            distances = []
            for j in range (4):
                distances.append(np.power(np.sum(np.multiply(np.subtract(i,random_centres[j]),np.subtract(i,random_centres[j]))),0.5))
                #distances.append(distance.euclidean(matrix[i],random_centres[j]))
            centroids[np.argmin(distances)].append(i)
            
        centres[0] = np.mean(centroids[0])
        centres[1] = np.mean(centroids[1])
        centres[2] = np.mean(centroids[2])
        centres[3] = np.mean(centroids[3])


        print(centres)
        
        centres = np.asarray(centres)
        
        if np.all(random_centres == centres):
            return centres
        random_centres = centres
```
